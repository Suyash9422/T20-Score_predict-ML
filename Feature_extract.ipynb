```python
df = pickle.load(open('dataset_level2.pkl','rb'))
```

```python
df.isnull().sum()
```

HERE WE CAN ABLE TO KNOW THAT IN CITY COLUMN 8548 CITY DATA IS NULL (MISSING)

```python
df[df['city'].isnull()]['venue'].value_counts()
```

HERE WE CAN OBSERVE THAT THE FIRST WORD OF VENUE REPRESENTS THE CITY THEREFORE WE CAN EXTARCT IT TO ANOTHER COLUMN 

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])  #HERE ARE COLLECTING THE FIRST WORD OF VENUE WHERE CITY IS NULL
#str.split results in seperating the words from the venue
#str.split().apply(lambda x:x[0]) results in collecting the first word from seperated words 
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

NOW WE CAN SEE THERE ARE NO CITIES WHERE THERE IS NULL VALUE OR MISSING DATA 

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

NOW WE ARE ONLY CONSIDERING THOSE MATCHES WHICH OCCURS IN A CITY MANY TIMES, SAY AROUND ONLY CONSIDERING THOSE CITIES WHERE MIN 5 MATCHES ARE PLAYED

```python
df['city'].value_counts() [df['city'].value_counts()> 600]
```

WE HAVE EXTRACTED THE LIST OF CITIES WHERE ATLEAST 5 MATCHES ARE PLAYED 

```python
eligible_cities = df['city'].value_counts() [df['city'].value_counts()> 600].index.tolist()
```

```python
eligible_cities
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df
```

we sepearated over and balls from ball column

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
```

```python
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)     #if there are some extras then value can be negative so we put condition that there some extras then consider 0 otherwise balls left
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

now we calculate current run rate 

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
groups = df.groupby('match_id')
match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
last_five
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df = final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

now there should be no bias so we are shuff;ing the data

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)
```

TRAINING OF DATA

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
